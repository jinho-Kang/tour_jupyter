```python
import folium
import googlemaps
import numpy as np
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
tour_street=pd.read_csv('data/서울시 관광거리 정보.csv',encoding='cp949')
tour_street.drop(['키','alias', '최종 표기명' ,'법정 시','법정 구','법정 동','행정 시','행정 구','행정 동'],axis=1,inplace=True)
```

```python
tour_street.rename(columns={tour_street.columns[2]:'경도'},inplace=True)
tour_street.rename(columns={tour_street.columns[3]:'위도'},inplace=True)
```

```python
tour_street
```

```python
f = open("key/동규key.txt", 'r')
gmaps_key =f.readlines()
gmaps = googlemaps.Client(key=gmaps_key[0])
```

```python
map = folium.Map(location=[tour_street['위도'].mean(), tour_street['경도'].mean()], zoom_start=11)

for n in tour_street.index:
    if np.isnan(tour_street['위도'][n]) == False: #df에 있는 lat가 nan이냐?
        folium.Marker([tour_street['위도'][n], tour_street['경도'][n]],
                      tooltip=tour_street['검색 키워드'][n],
                      popup = '<pre>' + tour_street['지번 주소'][n] + '</pre>').add_to(map)

map
```

```python
tour_street.rename(columns={'검색 키워드':'검색키워드'},inplace=True)
tour_street.rename(columns={'지번 주소':'지번주소'},inplace=True)
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine("mysql+pymysql://{user}:{pw}@localhost/{db}"
                       .format(user="root",
                               pw="1234",
                               db="tour"))
```

```python
tour_street.to_sql('tour_street', con = engine, if_exists = 'append', chunksize = 1000, index= False)
```
